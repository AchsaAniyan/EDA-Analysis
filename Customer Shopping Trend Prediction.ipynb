```python
# Importing all requried libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("shopping_trends.csv")
df
```

```python
df.rename(columns={'Item Purchased' : 'Item', 'Review Rating' : 'Rating'},inplace=True)
df.head()
```

```python
df.describe()
```

## Data Preprocessing

### Data Cleaning

```python
# Checking missing values
df.isnull().sum()
```

```python
# Checking duplicates
df.duplicated().sum()
```

```python
# Checking Datatypes
df.dtypes
```

```python
# Finding Categorical columns
cate_cols = [col for col in df.columns 
    if df[col].dtype == 'object']
cate_cols
```

```python
# Checking value counts of categorical columns
for i in cate_cols:
    print(i)
    print(df[i].value_counts())
```

### Data Encoding

```python
# label Encoding
le = LabelEncoder()
```

```python
le_cols=['Gender','Subscription Status','Discount Applied','Promo Code Used']
for i in le_cols:
    df[i] = le.fit_transform(df[i])
df.head()
```

* Gender : Male 1, Female 0
* Subscription Status : Yes 1, No 0
* Discount Applied : Yes 1, No 0
* PromoCode Used : Yes 1, No 0

### Handling Outliers

```python
sns.boxplot(df["Purchase Amount (USD)"])
```

## EDA

### Bivariate Analysis

```python
# PURCHASE AMOUNT WITH CATEGORY
sns.barplot(x='Category',y='Purchase Amount (USD)',data=df)
plt.title("Purchase Amount based on Category")
plt.show()
```

* Outerwear is least purchased

```python
# Purchase amount with and without Subscription
sns.barplot(x='Subscription Status',y='Purchase Amount (USD)',data=df)
plt.title("Purchase Amount Based on Subscription")
plt.show()
```

```python
sns.barplot(x='Payment Method',y='Purchase Amount (USD)',data=df)
plt.title("Most Commonly used payment method")
plt.show()
```

* The most commonly used payment method for purchases is credit card

```python
# Purchased amount based on season
sns.barplot(x='Season',y='Purchase Amount (USD)',data=df)
plt.title("payment method based on season")
plt.show()
```

* In Fall(Autum) season people more purchased

```python
sns.countplot(x="Season",hue='Category', data=df)
plt.title("Seasonal trends based on Category")
plt.show()
```

* In Spring and Winter Season Clothing are most purchased

```python
# Most used size based on Gender
sns.countplot(x="Size",hue='Gender', data=df)
plt.title("Size Preferences by Gender")
plt.show()
```

"M" sizeis mostly prefer by Men and Women

```python
# The mostly prefer Location by Gender
plt.figure(figsize=(10,6))
sns.countplot(x="Location",hue='Gender', data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(x="Promo Code Used",hue='Category', data=df)
plt.show()
```

* Most common Category purchased with promocode is clothing

```python
# The mostly prefer color in season wise
plt.figure(figsize=(10,6))
sns.countplot(x="Color",hue='Season', data=df)
plt.xticks(rotation=90)
plt.show()
```

* Silver color is most preferred in Summer season
